### 5. Univariate Analysis( Categorical Value Analysis -1 )

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('penguins.csv')
```

```python
df.shape
print(f"Total potential features :{df.shape[1]}")
print(f"Total sample data :{df.shape[0]}")
```

```python
df.head(7)
```

```python
df.sample(5)
```

```python
# missing value 
df.isnull().sum()
```

```python
# check duplicated value 
df.duplicated().sum()
```

```python
# permanently remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python

sns.countplot(data = df , x=df['species'],color = 'green')
```

```python
df['island'].value_counts()
```

```python
(df['island'].value_counts() / len(df) ) * 100
```

```python

island_count =  df['island'].value_counts()

island_label = df['island'].unique()

plt.pie(island_count, labels = island_label, autopct="%1.1f")
plt.title('Pie chart of Island ')
plt.show()
```

```python
df['sex'].value_counts()
```

```python
#use matplotlib

sex_count = df['sex'].value_counts()

sex_count = sex_count.sort_index() 

explode = [0.1] + [0] * (len(sex_count) - 1)

plt.pie(sex_count, labels=sex_count.index,
        autopct='%1.1f%%',explode =explode , shadow=True)
plt.title('Gender Distribution Pie Chart')
plt.show()
```

```python
# use seaborn 
Sex_count = df['sex'].value_counts()

plt.pie(Sex_count,
        labels=Sex_count.index,
        autopct="%1.1f%%")

plt.title('Gender Distribution Pie Chart')
plt.show()
```
